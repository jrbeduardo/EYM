```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Campo vectorial
def Positivo(x,y): 
    u = (x-900)/np.sqrt((x-900)**2 + y**2)
    v = y/np.sqrt((x-900)**2 + y**2)
    return u, v

def Negativo(x,y): 
    u = -(x+900)/np.sqrt((x+900)**2 + y**2)
    v = -y/np.sqrt((x+900)**2 + y**2)
    return u, v

```

```python
#Grid de puntos donde se coloca cada vector del campo vectorial
x,y = np.meshgrid(np.linspace(-1000,1000,20),np.linspace(-1000,1000,20))

#Se genera el campo vectorial para el grid
r,s = Negativo(x,y)
u,v = Positivo(x,y)

plt.rcParams["figure.figsize"] = (8,8)
plt.quiver(x,y,u+r,v+s)
plt.show()
```

```python
#Campo vectorial
def Campo(x,y): 
    u = -1/np.sqrt(1 + x**2)+1/np.sqrt(1 + y**2)
    v = -y/np.sqrt(1 + y**2)+x/np.sqrt(1 + x**2)
    return u, v
```

```python
#Grid de puntos donde se coloca cada vector del campo vectorial
x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))

#Se genera el campo vectorial para el grid
u,v = Campo(x,y)

plt.rcParams["figure.figsize"] = (8,8)
plt.quiver(x,y,u,v)
plt.show()
```
